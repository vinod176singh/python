```python
arr1 = np.random.randint(10,50,size = 10)
arr2 = np.random.randint(51,99, size = 10)

print("Array 1 >>", arr1)
print("Array 2 >>", arr2)
```

```python
arr3 = np.append(arr1,arr2) # axis is default i.e. None
print("Array 3 >>", arr3)
```

```python
arr4 = np.append(arr1,arr2, axis = 0) # axis is default i.e. None
print("Array 4 >>", arr4)
```

```python
arr5 = np.append(arr1,arr2, axis = 1) # axis is default i.e. None
print("Array 4 >>", arr5)
```

```python
## The axis is not available in 1 D array
```

#### Lets Try the Axis concept with 2D array

```python
arr1 = np.random.randint(10,50,size = (3,3))
arr2 = np.random.randint(51,99, size = (3,3))

print("Array 1 >>")
print(arr1)
print()
print("Array 2 >>")
print(arr2)
```

```python
arr3 = np.append(arr1,arr2) # Axis is default
print(arr3)
```

```python
arr4 = np.append(arr1,arr2, axis = 0) ## This will append rows of array2 to array1 
                                      ## and return a new array without making any changes to array1
print(arr4) # 6 x 3 array
```

```python
arr5 = np.append(arr1,arr2, axis = 1) ## This will append columns of array2 to array1 
                                      ## and return a new array without making any changes to array1
print(arr5) # 3 x 6
```

```python
 
    
    |---------------- axis(1) Column wise
    | [[10 11 12]
    |  [13 14 15]
    |  [16 17 18]]
    |
    |
  axis(0) Row wise
```

### 18. np.concatenate

```python
Syntax >>
np.concatenate(list_array,axis = 0)
np.concatenate([arr1,arr2,arr3....arrn], axis = 0)
```

```python
arr1 = np.random.randint(10,20, size = (3,3))
arr2 = np.random.randint(10,20, size = (3,3))
arr3 = np.random.randint(10,20, size = (3,3))

print("Array 1")
print(arr1)
print()
print("Array 2")
print(arr2)
print()
print("Array 3")
print(arr3)
print()
```

```python
arr4 = np.concatenate([arr1,arr2], axis = 0)
arr4
```

```python
arr4 = np.concatenate([arr1,arr2,arr3], axis = 0)
print(arr4) # 9 X 3
```

```python
arr5 = np.concatenate([arr1,arr2,arr3], axis = 1)
print(arr5) # 3 x 9
```

```python
AXIS 0 >>> VERTICAL AXIS
AXIS 1 >>> HORIZONTAL AXIS
```
