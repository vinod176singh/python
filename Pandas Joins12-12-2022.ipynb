### Concept of Inner Join and Outer Join

```python
## Inner >>> Intersection
## Outer >>> Union
```

```python
df1 = pd.DataFrame({"Student_Id":[1001,1002,1003,1004,1005],
                   "Student_Name":["Mangesh","Suresh","Dinesh","Ramesh","Jignesh"],
                   "Surname":["Patil","Kale","Wagh","Waghmare","Patel"],
                   'Age':[20,22,19,25,23]}, index=list("ABCDE"))

df2 = pd.DataFrame({"Student_Id":[1001,1002,1003,1004,1005],
                   "Student_Name":["Suresh","Dinesh","Ganesh","Kalpesh","Mangesh"],
                   "Surname":["Kale","Wagh","Bhamre","Joshi","Patil"],
                   'Age':[20,22,19,24,26],
                   "Location":["Amaravati","Pune","Satara","Akola","Mumbai"]}, index = list("ABCDF"))
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2], axis = 1)
```

```python
### Concat cares more about indices and not about the records itself
```

```python
pd.concat([df1,df2], axis = 1, join = "inner")
## Will join by axis = 1 i.e. columns and will look for comman indices in axis 0
```

```python
pd.concat([df1,df2], axis = 1, join = "outer") ## Default join in concat is outer
```

```python
### Inner join with concat along axis 0
pd.concat([df1,df2], axis = 0, join = "inner")
## Will concat by axis = 0 i.e. rows and will look for comman indices in axis 1 i.e. column
```

```python
df2.columns
```

```python
pd.concat([df1,df2], axis = 0, join = "inner")
```

```python
pd.concat([df1,df2], axis = 0, join = "outer")
```

```python
pd.concat([df1,df2], axis = 0) ### Outer is by default
```

### Pandas Merge

```python
Merge does not have concept of axis, rather it works on the concept of key
```

```python
df1 = pd.DataFrame({"Student_Id":[1001,1002,1003,1004,1005],
                   "Student_Name":["Mangesh","Suresh","Dinesh","Ramesh","Jignesh"],
                   "Surname":["Patil","Kale","Wagh","Waghmare","Patel"],
                   'Age':[20,22,19,25,23]})

df2 = pd.DataFrame({"Student_Id":[1005,1002,1003,1004,1001,1006,1007],
                   "Result":["Pass","Pass","Fail","Fail","Pass","Pass","Pass"],
                   "Percent":[88,89,43,63,56,90,95]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, how = "outer", on =["Student_Id"] )
```

```python
pd.merge(df1, df2, how = "inner", on =["Student_Id"] )
```

### Left_on, Right_on

```python
df1 = pd.DataFrame({"Student_Id":[1001,1002,1003,1004,1005],
                   "Student_Name":["Mangesh","Suresh","Dinesh","Ramesh","Jignesh"],
                   "Surname":["Patil","Kale","Wagh","Waghmare","Patel"],
                   'Age':[20,22,19,25,23]})

df2 = pd.DataFrame({"SiS_Id":[1005,1002,1003,1004,1001,1006,1007],
                   "Result":["Pass","Pass","Fail","Fail","Pass","Pass","Pass"],
                   "Percent":[88,89,43,63,56,90,95]})
```

```python
pd.merge(df1,df2,how = "outer",left_on = ["Student_Id"],right_on = ["SiS_Id"] )
```

```python
pd.merge(df1,df2,how = "inner",left_on = ["Student_Id"],right_on = ["SiS_Id"] )
```

```python
df_merged = pd.merge(df1,df2,how = "inner",left_on = ["Student_Id"],right_on = ["SiS_Id"] )
```

```python
df_merged 
```

```python
df2 = pd.DataFrame({"SiS_Id":[1005,1002,1003,1004,1001,1006,1007],
                   "PHY":[87,85,46,57,58,89,90],
                   "CHEM":[88,89,43,63,56,90,95],
                   "MATH":[88,90,42,67,60,85,92]})
```

```python
df_merged = pd.merge(df_merged,df2,how = "inner",left_on = ["Student_Id"],right_on = ["SiS_Id"] )
```
