```python
def clean_actors_list(x):
    x = x.replace("u'",'')
    x = x.replace("'","")
    x = re.sub("[^a-zA-Z,]"," ", x)
    x = [i.strip() for i in x.split(",")]
    return x
```

### Creating one column out of two

```python
array1 = np.random.randint(10,50, size=(10,5))
array1
```

```python
def get_total_two_columns(x,y):
    z = x + y
    return z
```

```python
df["A+B"] = df.apply(lambda x:get_total_two_columns(x["A"],x["B"]), axis = 1)
```

```python
name = ["John","Sarah","Krish","Pulkeet","Seema","Levina","Rajesh"]
surname = ["Hopkin","James","Roshan","Gupta","Chaturvedi","Kumar","Kuber"]
gender = ["Male","Female","Male","Male","Female","Female","Male"]
age = [20,"22","26",27,28,29,"23"]

df_1 = pd.DataFrame({"Name":name,"Surname":surname, "Gender":gender, "Age":age})
```

```python
df_1
```

```python
## Full Name With Salutations >> John Hopkins >> Mr. John Hopkin
```

```python
def get_full_name(name, surname, gender):
    if gender == "Male":
        return "Mr."+" "+name+" "+surname
    elif gender == "Female":
        return "Miss."+" "+name+" "+surname
```

```python
df_1["Full_Name"] = df_1.apply(lambda x: get_full_name(x["Name"],x["Surname"],x["Gender"]), axis = 1)
```

```python
df_1
```

### Creating two columns out of one

```python
df
```

```python
def get_column(col1,col2):
    val1 = col1+col2
    val2 = col1-col2
    val2 = abs(val2)
    return val1, val2
```

```python
df[["A+B+C", "A+B-C"]] = df.apply(lambda x: get_column(x["A+B"],x["C"]), axis = 1, result_type="expand")
```
