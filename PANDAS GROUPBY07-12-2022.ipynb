```python
df_new = pd.read_csv("HITRW.csv")
```

```python
df_new
```

```python
df_new["date"] = pd.to_datetime(df_new["date"])
```

```python
df_new["time"] = pd.to_datetime(df_new["time"])
```

```python
df_new.sort_values("time")
```

```python
### For aggregation and finding mean of the given dataset and for given frequency
df_new.groupby(pd.Grouper(key = "time", freq = "D")).mean()
```

```python
df_new.groupby(pd.Grouper(key = "time", freq = "W")).mean()
```

```python
df_new.groupby(pd.Grouper(key = "time", freq = "M")).mean()
```

```python
df_new.groupby(pd.Grouper(key = "time", freq = "7D")).sum()
```

```python
df_new.groupby(pd.Grouper(key = "time", freq = "D")).sum()["Vpmax"]
```

```python
df_new.groupby(pd.Grouper(key = "time", freq = "D")).sum()["Vpmax"].plot()
```
