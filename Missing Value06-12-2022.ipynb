```python
axis = 0 >> This will consider row indices
axis = 1 >> This will consider column indices
```

```python
df_new = pd.DataFrame({"A":[np.nan,2,3,np.nan,5,8,np.nan,np.nan,8], "B":[1,5,7,np.nan,np.nan,56,22,6,np.nan], "C":[11,22,33,44,55,66,77,88,99]})
```

```python
df_new
```

```python
df_new.dropna(axis = 0) ## Row Index 3rd and 4th has been removed 
```

```python
df_new.dropna(axis = 1)
```

```python
df_new.dropna(thresh=3,axis = 1) ## thresh = 3 means , if there are 3 non na values do not drop that column
```

```python
df_new.dropna(thresh=4,axis = 1) ## thresh = 4 means , if there are 4 non na values, then do not drop that column and if there
                                 ## are less than 4 non nan values drop the column
```

```python
df_new.dropna(thresh=5,axis = 1) ## thresh = 5 means , if there are 5 non na values, then do not drop that column and if there
                                 ## are less than 5 non nan values drop the column
```

```python
df_new.dropna(thresh=6,axis = 1) ## thresh = 6 means , if there are 6 non na values, then do not drop that column and if there
                                 ## are less than 6 non nan values drop the column
```

```python
df_new.dropna(thresh=1,axis = 0) ## thresh = 1 means , if there are 6 non na values, then do not drop that row and if there
                                 ## are less than 1 non nan values drop the row
```

```python
df_new.dropna(thresh=2,axis = 0) ## thresh = 1 means , if there are 6 non na values, then do not drop that row and if there
                                 ## are less than 1 non nan values drop the row
```

### Subset dropna

```python
## Normal dropna
df_new.dropna()
```

```python
## Drop na values with respect to column A only
df_new.dropna(subset = ["A"])
```

```python
## Drop na values with respect to column B only
df_new.dropna(subset = ["B"])
```
