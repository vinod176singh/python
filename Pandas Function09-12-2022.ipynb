### Iterating Over A Dataframe

```python
df = pd.DataFrame(np.random.randint(100,200,size = (10,4)), columns = list("ABCD"))
```

```python
df
```

### 1. df.items()

```python
for col_name, col_data, in df.items():  ## df.items returns col_name and col_data
    print(col_name,">>>>")
    print(col_data)
```

```python
for col_name, col_data, in df.items():  ## df.items returns col_name and col_data
    print(col_name,">>>>")
#     print(col_data)
```

```python
df.items
```

### df.iteritems()

```python
for col_name, col_data in df.iteritems():
    print(col_name)
```

```python
for col_name, col_data in df.iteritems():
    print(col_name)
    print(col_data)
    break
```

```python
type(col_data)
```

```python
for col_name, col_data in df.iteritems():
    print(col_name,">>>",col_data.to_list())
    
```

```python
df.iteritems
```

### Iterating over Row

```python
for row_index in df.index:
    print(row_index," ", end = " ")
```

```python
for row_index in df.index:
    print("Row Index >> ", row_index)
    print(df.loc[row_index])
    break
```

```python
for row_index, row_data in df.iterrows(): ###
    print("Row Index >>> ",row_index)
    print(row_data)
    print()
```

```python
for row_index, row_data in df.iterrows():
    print(row_index)
    print(row_data.to_list())
```
