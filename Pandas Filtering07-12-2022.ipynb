## Filtering Using LOC

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("movies_data.csv")
```

```python
df.head()
```

```python
df["content_rating"].unique()
```

```python
df.genre.unique()
```

```python
df.star_rating.unique()
```

```python
df.loc[df.content_rating=="R"]  ## Using loc
                                ## Using loc we can assign value to the given location condition
```

```python
df.loc[df["content_rating"]=="R"] 
```

```python
df[df.content_rating == "R"] ## Without loc
```

```python
df.loc[df["content_rating"]=="PG"]
```

```python
df[df["content_rating"]=="PG"]
```

```python
df.loc[df["genre"]=="Action"]
```

```python
df.loc[(df["genre"]=="Action")&(df['content_rating']=="PG")]
```

```python
df[(df["genre"]=="Action")&(df['content_rating']=="PG")]
```

```python
df[df["duration"]>150]
```

```python
df[(df["duration"]>150)&(df["genre"]=="Action")]
```

```python
df[(df["duration"]>150)&(df["genre"]=="Action")&(df["star_rating"]>=9)]
```

```python
df[(df["duration"]>150)&((df["genre"]=="Action")|(df["genre"]=="Crime"))]
```

```python
df[(df["duration"]>150)|(df["genre"]=="Comedy")]
```

```python
df_action = df[df["genre"]=="Action"]
```
